```python
import pandas as pd
```

```python
cantidad = 1000000
```

```python
import random
```

```python
#Datos aleatorios
influencer = ["AnaSof", "Ricardo Milos", "El Borrego", "Paco Web", "Luperra", "Domelipa", "Arquijuve", "Lupitics"]
comentarios = ["Está horrible, me encanta", "Playa Delfines la mejor", "Vayan a CocoBongo", "CocoBongo sobrevalorado", "Apesta a pescado en la noche", "Me robé arena aunque era propiedad privada", "Hola mundo", "Vayan Boston, las mejores pizza-hamburguesas"]
tipo_post = ["foto", "video", "comentario"]
```

```python
#Creacion del DataFrame
dfInfluencer = pd.DataFrame ({
    "influencer" : [random.choice(influencer) for _ in range(cantidad)],
    "likes": [round(random.uniform(5000,500000),2) for _ in range(cantidad)],
    "comentarios" : [random.choice(comentarios) for _ in range(cantidad)],
    "tipo_post" : [random.choice(tipo_post) for _ in range(cantidad)],
    "fecha" : [random.randint(2020,2026) for _ in range(cantidad)]
})
```
