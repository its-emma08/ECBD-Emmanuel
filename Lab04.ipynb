```python
#Limpieza
#Nombres de columnas en minusculas
df.columns = df.columns.str.strip().str.lower()
```

```python
df.columns
```

```python
#Detectar valores nulos 
df.isnull().sum()
```

```python
#Eliminar columna 
df.drop(columns=["cabin"], inplace=True)
```

```python
#Rellenar edades con una edad promedio 
df["age"] = df["age"].fillna(df["age"].median())
```

```python
df.isnull().sum()
```

```python
#Rellenamos la columna Fare con el valor más repetido (moda) 
df["fare"] = df["fare"].fillna(df["fare"].mode()[0])
```

```python
df.isnull().sum()
```

```python
#verificar valores duplicados 
df.duplicated().sum()
```

```python
#Validar datos
df["age"].min()
```

```python
df["age"].max()
```

```python
#Graficamos el total de pasajeros por clase
plt.figure(figsize=(18,4)) #tamaño
sns.countplot(df, x="pclass")
plt.title("Total de pasajeros por clase")
plt.xlabel("Clase")
plt.show()
```

```python
#Graficamos las clases por genero 
plt.figure(figsize=(8,4))
sns.countplot(df, x="sex", hue="pclass")
plt.title("Total de genero por Clases")
plt.xlabel("Genero")
plt.ylabel("Total")
plt.show()
```

```python
#Mostramos la edad por Genero y clase 
plt.figure(figsize=(6,4))
sns.violinplot(df, x="sex", y="age", hue="pclass")
plt.title("Rango de edad por género y clase")
plt.xlabel("Genero")
plt.ylabel("Edad")
plt.show()
```

```python
#CONOCIMIENTO 
#Los pasajeros de primera claser pagaban tarifas más altas.
df.groupby("pclass")["fare"].mean()
```
